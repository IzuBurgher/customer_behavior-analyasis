```python
import pandas as pd

df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
#let's replace the missing review item with the median value of the same category

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#fixing the naming convention to a snakecase format for consistency
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amounts'})
```

```python
df.columns
```

```python
#Creating new columns age_group
#qcut categories the column into specified equal parts like q4 or q5

labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
#Creating column purchase_frequency_days
#Converting frequency of purchases into numbers to allow easy analysis
#Creating a dictionary called frequency mapping

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases', 'purchase_frequency_days']].head(10)
```

```python
#More analsis: checking if these 2 columns are the same

df[['discount_applied', 'promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis=1)
```
